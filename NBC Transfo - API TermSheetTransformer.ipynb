# Guide des fonctions `TermSheetTransformer`

Ce notebook présente les principales fonctions disponibles dans la classe `TermSheetTransformer` (fichier `ts_editor_D.py`) ainsi que des exemples d'utilisation typiques.

Objectif : avoir un **cheat sheet exécutable** pour préparer et tester les transformations de term sheets NBC.

---

## 1. Préparation : import & chargement du document


```python
from ts_editor_D import TermSheetTransformer

# Chemin vers ton document Word (.docx ou .doc)
CHEMIN_DOCUMENT = "ts exemple.docx"  # à adapter

# Création du transformer
transformer = TermSheetTransformer(CHEMIN_DOCUMENT)
transformer
```

## 2. Sections du tableau principal

Fonctions disponibles :

- `replace_word(old, new, occurrence=1)`
- `set_section(title, description, after_section=None, red_highlight_in_final=False, occurrence=1)`
- `add_section_after(after_section, title, description, occurrence=1)`
- `update_section_description(section_title, new_description, occurrence=1)`
- `set_section_order(section_order)`
- `remove_section(section_title, occurrence=1)`

Les cellules suivantes montrent des exemples typiques.


```python
# 2.1 Remplacer du texte partout (tableau + paragraphes)
transformer.replace_word("Notes", "Securities")
transformer.replace_word("Certificates", "Securities")

# 2.2 Ajouter ou mettre à jour une section (si existe -> modif, sinon -> création)
transformer.set_section(
    title="Issuer",
    description="National Bank of Canada"
)

# 2.3 Ajouter une section explicitement après une autre (toujours ajout)
transformer.add_section_after(
    after_section="Listing",
    title="New Section After Listing",
    description="Description de la nouvelle section"
)

# 2.4 Mettre à jour uniquement si la section existe (sinon ne fait rien)
transformer.update_section_description(
    section_title="Calculation Agent",
    new_description="National Bank of Canada"
)

# 2.5 Supprimer une section
transformer.remove_section("Guarantor")

# 2.6 Réordonner les sections
transformer.set_section_order([
    "Issue Type",
    "Issuer",
    "Series Number",
    "Dealer",
    "Currency",
    "Trade Date",
    "Maturity Date",
    "Final Redemption",
    "Governing Law",
])
```

## 3. Paragraphes hors tableau (texte libre)

Fonctions disponibles :

- `add_content(text, after_paragraph=None, red_highlight_in_final=False)`
- `update_paragraph(text_contains, new_text, occurrence=1)`
- `remove_paragraph(text_contains, occurrence=1)`


```python
# 3.1 Ajouter du texte libre à la fin
transformer.add_content(
    text="Important notice: This term sheet is a summary only.",
    red_highlight_in_final=True
)

# 3.2 Ajouter après un paragraphe existant
transformer.add_content(
    text="Additional information for investors.",
    after_paragraph="Investors should be aware that"
)

# 3.3 Modifier un paragraphe existant (si trouvé)
transformer.update_paragraph(
    text_contains="Gross Total Return",
    new_text="Gross Total Return (updated definition)"
)

# 3.4 Supprimer un paragraphe
transformer.remove_paragraph(
    text_contains="This paragraph is no longer applicable."
)
```

## 4. Sections de Disclaimer (après les tableaux)

Fonctions disponibles :

- `set_disclaimer_section(title, content, after_title=None, red_highlight_in_final=False, occurrence=1)`
- `update_disclaimer_content(title, new_content, red_highlight_in_final=False, occurrence=1)`
- `add_disclaimer_content(title, additional_content, red_highlight_in_final=False, occurrence=1)`
- `remove_disclaimer_section(title, occurrence=1)`


```python
# 4.1 Créer ou mettre à jour une section de disclaimer
transformer.set_disclaimer_section(
    title="IMPORTANT INFORMATION",
    content=(
        "This term sheet is a summary of the terms of the Securities.\n"
        "Investors should read the Base Prospectus."
    )
)

# 4.2 Ajouter une nouvelle section après un titre existant
transformer.set_disclaimer_section(
    title="Additional Risk Factors",
    content="Investors may lose part or all of their invested capital.",
    after_title="IMPORTANT INFORMATION",
    red_highlight_in_final=True
)

# 4.3 Mettre à jour uniquement le contenu si la section existe
transformer.update_disclaimer_content(
    title="Additional Risk Factors",
    new_content=(
        "Investors may lose part or all of their invested capital.\n"
        "The product is intended for knowledgeable investors only."
    ),
    red_highlight_in_final=True
)

# 4.4 Ajouter du contenu à une section existante
transformer.add_disclaimer_content(
    title="IMPORTANT INFORMATION",
    additional_content="The information contained herein is subject to change without notice."
)

# 4.5 Supprimer une section de disclaimer complète
transformer.remove_disclaimer_section("Old Disclaimer Section")
```

## 5. Logo & exécution

- `add_logo(logo_path, width_inches=1.0, all_sections=True)`
- `execute_and_export(output_dir, markup_docx, final_docx, final_pdf)`


```python
# 5.1 Ajout du logo dans le header
transformer.add_logo(
    logo_path="logo_nbc.png",  # chemin vers ton logo
    width_inches=1.6,
    all_sections=True
)

# 5.2 Exécution et export des 3 versions
transformer.execute_and_export(
    output_dir="./output",
    markup_docx="modifications.docx",
    final_docx="document_final.docx",
    final_pdf="document_final.pdf"
)
```
